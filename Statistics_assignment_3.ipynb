```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Z Statistic = {z:.4f}")
    print(f"P-Value = {p_value:.4f}")

    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

if __name__ == "__main__":
    sample = [88, 92, 85, 91, 87, 90, 89, 94, 86, 90]

    population_mean = 85
    population_std = 4.5

    z_test(sample, population_mean, population_std)
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):

    sample_data = np.array(sample_data)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    if p_value < alpha:
        result = "Reject the null hypothesis: significant difference."
    else:
        result = "Fail to reject the null hypothesis: no significant difference."

    return z_stat, p_value, result


if __name__ == "__main__":
    np.random.seed(0)
    sample = np.random.normal(loc=102, scale=15, size=40)

    population_mean = 100
    population_std = 15

    z, p, interpretation = one_sample_z_test(sample, population_mean, population_std)

    print(f"Z Statistic = {z:.4f}")
    print(f"P-Value = {p:.4f}")
    print(f"Conclusion: {interpretation}")
```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    z_critical = norm.ppf(1 - alpha / 2)

    if abs(z_stat) > z_critical:
        result = "Reject the null hypothesis (significant difference)"
    else:
        result = "Fail to reject the null hypothesis (no significant difference)"

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z Statistic = {z_stat:.4f}")
    print(f"P-Value = {p_value:.4f}")
    print(f"Z Critical (±{alpha/2:.2f} significance): ±{z_critical:.4f}")
    print(f"Conclusion: {result}")

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (Left)')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (Right)')

    plt.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Z Statistic = {z_stat:.2f}')
    plt.axvline(-z_critical, color='black', linestyle=':', label=f'-Z Critical = {-z_critical:.2f}')
    plt.axvline(z_critical, color='black', linestyle=':', label=f'+Z Critical = {z_critical:.2f}')

    plt.title('Two-Tailed Z-Test Visualization')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    np.random.seed(42)
    sample_data = np.random.normal(loc=102, scale=15, size=50)

    population_mean = 100
    population_std = 15
    alpha = 0.05

    two_tailed_z_test_with_plot(sample_data, population_mean, population_std, alpha)
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0, mu_1, sigma, n, alpha=0.05, tail='right'):

    se = sigma / np.sqrt(n)

    if tail == 'right':
        z_crit = norm.ppf(1 - alpha)
        x_crit = mu_0 + z_crit * se
        beta = norm.cdf((x_crit - mu_1) / se)

    elif tail == 'left':
        z_crit = norm.ppf(alpha)
        x_crit = mu_0 + z_crit * se
        beta = 1 - norm.cdf((x_crit - mu_1) / se)

    elif tail == 'two':
        z_crit_low = norm.ppf(alpha / 2)
        z_crit_high = norm.ppf(1 - alpha / 2)
        x_crit_low = mu_0 + z_crit_low * se
        x_crit_high = mu_0 + z_crit_high * se
        beta = norm.cdf((x_crit_high - mu_1) / se) - norm.cdf((x_crit_low - mu_1) / se)

    else:
        raise ValueError("tail must be 'right', 'left', or 'two'")

    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    y0 = norm.pdf(x, mu_0, se)
    y1 = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀ Distribution (μ₀)', color='blue')
    plt.plot(x, y1, label='H₁ Distribution (μ₁)', color='green')

    if tail == 'right':
        plt.fill_between(x, 0, y0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, 0, y1, where=(x < x_crit), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')

    elif tail == 'left':
        plt.fill_between(x, 0, y0, where=(x <= x_crit), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, 0, y1, where=(x > x_crit), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')

    elif tail == 'two':
        plt.fill_between(x, 0, y0, where=(x <= x_crit_low) | (x >= x_crit_high), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, 0, y1, where=(x > x_crit_low) & (x < x_crit_high), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(x_crit_low, color='black', linestyle='--', label='Critical Values')
        plt.axvline(x_crit_high, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")
mu_0 = 100
mu_1 = 105
sigma = 15
n = 36
alpha = 0.05

visualize_type1_type2_errors(mu_0, mu_1, sigma, n, alpha, tail='right')

```

```python
# 6: Write a Python program to perform an independent T-test and interpret the results

from scipy.stats import ttest_ind

def perform_independent_t_test(group1, group2, alpha=0.05):

  t_statistic, p_value = ttest_ind(group1, group2)

  print(f"T-statistic: {t_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < alpha:
    return "The p-value is less than the significance level, so we reject the null hypothesis. There is a statistically significant difference between the two groups."
  else:
    return "The p-value is greater than the significance level, so we fail to reject the null hypothesis. There is no statistically significant difference between the two groups."


group_a = [15, 18, 12, 17, 19, 16, 14, 20, 13, 15]
group_b = [22, 25, 20, 23, 24, 21, 26, 27, 22, 24]

interpretation = perform_independent_t_test(group_a, group_b)
interpretation
```

```python
# 7: Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

def paired_t_test_with_visualization(data_before, data_after, alpha=0.05):

  t_statistic, p_value = ttest_rel(data_before, data_after)

  print(f"T-statistic: {t_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < alpha:
    print(
        "The p-value is less than the significance level, so we reject the null hypothesis."
        " There is a statistically significant difference between the two groups."
    )
  else:
    print(
        "The p-value is greater than the significance level, so we fail to reject the null hypothesis."
        " There is no statistically significant difference between the two groups."
    )

  plt.figure(figsize=(8, 6))
  plt.scatter(data_before, data_after, alpha=0.7)
  plt.xlabel("Data Before")
  plt.ylabel("Data After")
  plt.title("Paired Sample T-Test Visualization")
  plt.plot([min(data_before), max(data_before)],
           [min(data_before), max(data_before)],
           color="red", linestyle="--", label="No Change Line")
  plt.legend()
  plt.show()


data_before = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
data_after = [12, 14, 17, 20, 22, 25, 28, 30, 33, 35]

paired_t_test_with_visualization(data_before, data_after)
```

```python
# 8: Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
from scipy.stats import norm

np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=50)
group2 = np.random.normal(loc=12, scale=2, size=50)

population_std1 = 2
population_std2 = 2
z_statistic = (np.mean(group1) - np.mean(group2)) / np.sqrt(population_std1**2 / len(group1) + population_std2**2 / len(group2))
p_value_z = 2 * (1 - norm.cdf(abs(z_statistic)))

t_statistic, p_value_t = ttest_ind(group1, group2)

print(f"Z-Test: Z-statistic = {z_statistic:.4f}, P-value = {p_value_z:.4f}")
print(f"T-Test: T-statistic = {t_statistic:.4f}, P-value = {p_value_t:.4f}")


if p_value_z < 0.05 and p_value_t < 0.05:
  print("Both Z-test and T-test indicate a statistically significant difference between the groups.")
elif p_value_z >= 0.05 and p_value_t >= 0.05:
  print("Both Z-test and T-test indicate no statistically significant difference between the groups.")
else:
  print("Z-test and T-test produce conflicting results.")

before_treatment = np.random.normal(loc=15, scale=3, size=30)
after_treatment = before_treatment + np.random.normal(loc=2, scale=1, size=30)

t_statistic_paired, p_value_paired = ttest_rel(before_treatment, after_treatment)

print("\nPaired T-Test:")
print(f"T-statistic = {t_statistic_paired:.4f}, P-value = {p_value_paired:.4f}")

if p_value_paired < 0.05:
  print("The paired T-test indicates a statistically significant difference between the pre-treatment and post-treatment groups.")
else:
  print("The paired T-test indicates no statistically significant difference between the pre-treatment and post-treatment groups.")
```

```python
# 9: rite a Python function to calculate the confidence interval for a sample mean and explain its significance. Java + DS

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence_level=0.95):

  n = len(data)
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof=1)
  standard_error = sample_std / np.sqrt(n)
  t_critical = t.ppf((1 + confidence_level) / 2, df=n - 1)
  margin_of_error = t_critical * standard_error
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error
  return lower_bound, upper_bound

sample_data = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
confidence_level = 0.95

lower, upper = confidence_interval(sample_data, confidence_level)
print(f"The {confidence_level * 100}% confidence interval for the sample mean is: ({lower:.2f}, {upper:.2f})")


# Significance of the Confidence Interval

# 1. Estimation of the Population Mean:
#    - A confidence interval provides a range of plausible values for the true population mean.
#    - It's more informative than simply calculating a point estimate (the sample mean) because it shows the uncertainty associated with that estimate.

# 2. Inferring Population Parameters:
#    - By constructing a confidence interval, we can infer the likely range of values that the population mean falls within.
#    - This allows us to make more informed decisions and draw conclusions about the population based on our sample data.

# 3. Assessing the Precision of the Sample Mean:
#    - The width of the confidence interval reflects the precision of our estimate of the population mean.
#    - A narrower interval indicates a more precise estimate. Factors like sample size and variability within the data influence the interval width.

# 4. Hypothesis Testing:
#    - Confidence intervals can be used to indirectly test hypotheses.
#    - For example, if a hypothesized value of the population mean falls outside the confidence interval, we would reject the null hypothesis that the population mean equals that value.

# 5. Communication of Uncertainty:
#    - When reporting results based on sample data, it's essential to clearly communicate the uncertainty associated with our findings.
#    - The confidence interval provides a standardized way to convey this uncertainty to others.
```

```python
# 10: Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence_level=0.95):

  n = len(data)
  sample_std = np.std(data, ddof=1)
  standard_error = sample_std / np.sqrt(n)
  t_critical = t.ppf((1 + confidence_level) / 2, df=n - 1)
  margin_of_error = t_critical * standard_error
  return margin_of_error

sample_data = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
confidence_level = 0.95

error = margin_of_error(sample_data, confidence_level)
print(f"The margin of error for the sample mean at a {confidence_level * 100}% confidence level is: {error:.2f}")
```

```python
# 11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np
from scipy.stats import beta

def bayesian_inference(prior_a, prior_b, data_successes, data_failures):


  posterior_a = prior_a + data_successes
  posterior_b = prior_b + data_failures
  return posterior_a, posterior_b

prior_a = 2
prior_b = 2

data_successes = 7
data_failures = 3

posterior_a, posterior_b = bayesian_inference(prior_a, prior_b, data_successes, data_failures)

print(f"Prior Distribution: Beta({prior_a}, {prior_b})")
print(f"Posterior Distribution: Beta({posterior_a}, {posterior_b})")


# Process Explanation:

# 1. Prior Belief:
#    - We start with a prior belief about the parameter we want to estimate.
#    - In this example, we are estimating the probability of success (e.g., flipping a coin and getting heads).
#    - We represent this prior belief using a probability distribution. The Beta distribution is often used for representing probabilities.

# 2. Data Observation:
#    - We collect data (observations) related to the parameter we want to estimate.
#    - In this example, we flipped the coin 10 times and observed the outcomes (heads or tails).

# 3. Likelihood Function:
#    - The likelihood function describes how likely our observed data is given different values of the parameter we're estimating.
#    - For example, the likelihood of getting 7 heads and 3 tails in 10 flips depends on the underlying probability of getting a head.

# 4. Bayes' Theorem:
#    - We use Bayes' Theorem to update our prior belief (prior distribution) in light of the observed data (likelihood).
#    - Bayes' Theorem calculates the posterior distribution, which represents our updated belief about the parameter after observing the data.

# 5. Posterior Distribution:
#    - The posterior distribution is the result of combining the prior belief with the likelihood function.
#    - It represents our updated belief about the parameter, taking into account both our prior knowledge and the new data.


# In the example code, we used the Beta distribution to represent both the prior and posterior distributions.
# The update involves adding the number of successes and failures from the data to the parameters of the Beta distribution.
# The posterior distribution is more concentrated around a specific probability than the prior, reflecting the influence of the data.
```

```python
#12: Perform a Chi-square test for independence between two categorical variables in Python
import pandas as pd
from scipy.stats import chi2_contingency
data = {'Tea': [30, 10], 'Coffee': [20, 40]}
contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

print("Contingency Table:")
print(contingency_table)
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi2 Statistic: {chi2}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p}")
print("\nExpected Frequencies Table:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))
```

```python
#14 : Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

from scipy.stats import chisquare

def goodness_of_fit_test(observed_data, expected_distribution):

  chi2, p_value = chisquare(observed_data, expected_distribution)
  return chi2, p_value


observed_data = [15, 20, 25, 22, 18]
expected_distribution = [20, 20, 20, 20, 20]

chi2, p_value = goodness_of_fit_test(observed_data, expected_distribution)
print("Chi-square statistic:", chi2)
print("P-value:", p_value)

if p_value < 0.05:
  print("The observed data significantly differs from the expected distribution.")
else:
  print("The observed data does not significantly differ from the expected distribution.")
```

```python
# 16: Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

def f_test_variances(data1, data2, alpha=0.05):


  variance1 = np.var(data1, ddof=1)
  variance2 = np.var(data2, ddof=1)

  f_statistic = variance1 / variance2

  df1 = len(data1) - 1
  df2 = len(data2) - 1

  p_value = f.sf(f_statistic, df1, df2)

  print(f"F-statistic: {f_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < alpha:
    return "The p-value is less than the significance level, so we reject the null hypothesis. There is a statistically significant difference between the variances of the two groups."
  else:
    return "The p-value is greater than the significance level, so we fail to reject the null hypothesis. There is no statistically significant difference between the variances of the two groups."

sample1 = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
sample2 = [12, 14, 17, 20, 22, 25, 28, 30, 33, 35]

interpretation = f_test_variances(sample1, sample2)
interpretation
```

```python
# 17: Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import numpy as np
import pandas as pd
from scipy.stats import f_oneway

def perform_anova_test(data):

  groups = list(data.values())
  f_statistic, p_value = f_oneway(*groups)

  print(f"F-statistic: {f_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if p_value < 0.05:
    return "The p-value is less than the significance level, so we reject the null hypothesis. There is a statistically significant difference in means between at least two of the groups."
  else:
    return "The p-value is greater than the significance level, so we fail to reject the null hypothesis. There is no statistically significant difference in means between the groups."

data = {
    'Group A': [10, 12, 15, 18, 20],
    'Group B': [12, 14, 17, 20, 22],
    'Group C': [8, 10, 11, 13, 15]
}

interpretation = perform_anova_test(data)
interpretation
```

```python
# 19: Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(data):

  results = {}

  for group_name, group_data in data.items():
    stat, p = shapiro(group_data)
    results[f"Normality ({group_name})"] = {
        "statistic": stat,
        "p-value": p,
        "interpretation": "Not normal" if p < 0.05 else "Normal"
    }

  stat, p = levene(*data.values())
  results["Equal Variance"] = {
      "statistic": stat,
      "p-value": p,
      "interpretation": "Unequal variances" if p < 0.05 else "Equal variances"
  }

  results["Independence"] = {
      "interpretation": "Assumption not automatically tested. Needs to be assessed based on study design."
  }

  return results


data = {
    'Group A': [10, 12, 15, 18, 20],
    'Group B': [12, 14, 17, 20, 22],
    'Group C': [8, 10, 11, 13, 15]
}

assumption_results = check_anova_assumptions(data)
assumption_results
```

```python
# 24: Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
from scipy.stats import chi2

data = [14, 15, 16, 14, 15, 15, 16, 14, 13, 15]
sample_size = len(data)
sample_var = np.var(data, ddof=1)

sigma_squared_0 = 1.5

chi2_stat = (sample_size - 1) * sample_var / sigma_squared_0

df = sample_size - 1

alpha = 0.05

chi2_crit_low = chi2.ppf(alpha / 2, df)
chi2_crit_high = chi2.ppf(1 - alpha / 2, df)

print(f"Chi-square stat: {chi2_stat:.3f}")
print(f"Critical values: ({chi2_crit_low:.3f}, {chi2_crit_high:.3f})")

if chi2_stat < chi2_crit_low or chi2_stat > chi2_crit_high:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

```

```python
# 25: Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
from scipy.stats import norm

def z_test_proportions(p1, n1, p2, n2, alternative='two-sided'):


  p_pooled = (p1 * n1 + p2 * n2) / (n1 + n2)
  se = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n1 + 1 / n2))
  z_statistic = (p1 - p2) / se

  if alternative == 'two-sided':
    p_value = 2 * norm.sf(abs(z_statistic))
  elif alternative == 'less':
    p_value = norm.cdf(z_statistic)
  elif alternative == 'greater':
    p_value = norm.sf(z_statistic)
  else:
    raise ValueError("Invalid alternative hypothesis specified.")

  result = "Reject null hypothesis" if p_value < 0.05 else "Fail to reject null hypothesis"

  return {'z_statistic': z_statistic, 'p_value': p_value, 'result': result}

p1 = 0.6
n1 = 100
p2 = 0.4
n2 = 100

test_result = z_test_proportions(p1, n1, p2, n2)

print("Z-statistic:", test_result['z_statistic'])
print("P-value:", test_result['p_value'])
print("Result:", test_result['result'])
```

```python
# 26: Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

def f_test_variances_with_visualization(data1, data2, alpha=0.05):


  variance1 = np.var(data1, ddof=1)
  variance2 = np.var(data2, ddof=1)

  f_statistic = variance1 / variance2

  df1 = len(data1) - 1
  df2 = len(data2) - 1

  p_value = f.sf(f_statistic, df1, df2)

  print(f"F-statistic: {f_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  x = np.linspace(0, max(f_statistic * 2, 5), 500)
  y = f.pdf(x, df1, df2)
  plt.plot(x, y)
  plt.title(f"F-distribution (dfn={df1}, dfd={df2})")
  plt.xlabel("F-statistic")
  plt.ylabel("Probability Density")
  plt.axvline(x=f_statistic, color='red', linestyle='--', label='Calculated F-statistic')
  plt.legend()
  plt.show()

  if p_value < alpha:
    return "The p-value is less than the significance level, so we reject the null hypothesis. There is a statistically significant difference between the variances of the two groups."
  else:
    return "The p-value is greater than the significance level, so we fail to reject the null hypothesis. There is no statistically significant difference between the variances of the two groups."

sample1 = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
sample2 = [12, 14, 17, 20, 22, 25, 28, 30, 33, 35]

interpretation = f_test_variances_with_visualization(sample1, sample2)
interpretation
```
